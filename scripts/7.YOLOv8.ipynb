6. Train YOLOv8-L

```python
# Load COCO-pretrained YOLOv8-large
from ultralytics import YOLO
model = YOLO("yolov8l.pt")

# Train — key arguments explained inline
model.train(
    data="yolo_dataset/data.yaml",  # Path to YAML describing train/val dirs and class names/count
    epochs=50,                      # Full passes over the training set
    imgsz=640,                      # Resize to 640×640 for speed/compatibility
    batch=32,                       # Images per step (increase if GPU memory allows)
    device=0,                       # GPU index; use 'cpu' or -1 for CPU
    project="runs/detect",          # Root folder where runs are saved
    name="traffic_light",           # Subfolder name for this run
    exist_ok=False,                 # Prevent overwriting an existing run folder
    augment=False,                  # Built-in augmentations (Mosaic/MixUp/HSV etc.); set True to enable
    patience=10,                    # Early stop if no val improvement for N epochs
)
```
